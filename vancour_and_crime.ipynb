```r
crime_data <- read_csv("C:\\Users\\Kaiyan Zhang\\Desktop\\econ326\\crimedata_csv_AllNeighbourhoods_AllYears\\crimedata_csv_AllNeighbourhoods_AllYears.csv")
```

```r
distinct(crime_data, TYPE)
```

```r
crime_data_selected <- crime_data|>
    filter(TYPE == c("Vehicle Collision or Pedestrian Struck (with Fatality)",
    "Vehicle Collision or Pedestrian Struck (with Injury)",
    "Break and Enter Commercial",
    "Break and Enter Residential/Other",
    "Homicide"))|>
    select(TYPE, YEAR, NEIGHBOURHOOD)

glimpse(crime_data_selected)
```

```r
crime_data_selected <- crime_data_selected|>
    group_by(TYPE)|>
    arrange(YEAR)|>
    rename(type = TYPE, year = YEAR, neighbourhood = NEIGHBOURHOOD)

head(crime_data_selected)

```
